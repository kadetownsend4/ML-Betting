```python
import pandas as pd
import nfl_data_py as nfl

# import the weekly data for each player in the 2024 NFL season 
weekly_stats_2024 = nfl.import_weekly_data([2024])
weekly_stats_2024 = nfl.clean_nfl_data(weekly_stats_2024)

# Create a filtered copy with relevant stats for passing & rushing
qb_df = weekly_stats_2024[weekly_stats_2024['position'] == 'QB'][['player_id', 'player_name', 'player_display_name','position', 'recent_team', 'week', 'season_type',
                           'opponent_team', 'passing_yards',
                           'passing_tds',  'sacks', 'sack_yards', 'passing_air_yards', 'passing_yards_after_catch',
                           'passing_first_downs', 'passing_epa', 'passing_2pt_conversions', 'pacr',
                            'rushing_yards', 'rushing_tds','fantasy_points', 'fantasy_points_ppr']].copy()

# Checking to see if the list contains only quarterbacks 
print(qb_df['player_display_name'].unique()) 

# display the weekly data
display(qb_df)
```

```python
# Get weekly passing stats for 2024
weekly_passing = nfl.clean_nfl_data(nfl.import_weekly_pfr(s_type="pass", years=[2024]))
weekly_passing = weekly_passing.drop(['game_id', 'pfr_game_id', 'game_type','team','opponent','pfr_player_id','season'], axis=1)
# print(weekly_passing.columns)

# # Checking to see if the list contains only quarterbacks 
# print(weekly_passing['pfr_player_name'].unique()) 

# Get weekly receiving stats for 2024
weekly_rushing = nfl.clean_nfl_data(nfl.import_weekly_pfr(s_type="rush", years=[2024]))
weekly_rushing = weekly_rushing.drop(['game_id', 'pfr_game_id', 'game_type','team','opponent','pfr_player_id','season'], axis=1)
# print(weekly_rushing.columns)

# # Display first few rows
# display(weekly_passing.head())
# display(weekly_rushing.head())


qb_rushing_and_passing_weekly = pd.merge(weekly_passing, weekly_rushing, on=['pfr_player_name', 'week'], how='inner')

qb_rushing_and_passing_weekly['pfr_player_name'] = qb_rushing_and_passing_weekly['pfr_player_name'].replace('Gardner Minshew II', 'Gardner Minshew')
players_to_remove = [
    'Malik Nabers', 'Ricky Pearsall', 'Adonai Mitchell', 'D.J. Moore', 
    'Taysom Hill', 'Dyami Brown', 'David Montgomery', 
    'Jameson Williams', 'Joe Mixon'
]
# Checking to see if the list contains only quarterbacks 
print(qb_rushing_and_passing_weekly['pfr_player_name'].unique()) 

# https://chatgpt.com/share/67c0f7b6-8950-800f-a640-2bd9b595cf13 
# I asked chat gpt the easiest way to remove the players which were in one dataframe, but not the other

# Filter out the specified players
qb_rushing_and_passing_weekly = qb_rushing_and_passing_weekly[
    ~qb_rushing_and_passing_weekly['pfr_player_name'].isin(players_to_remove)
]


# diff2 = list(set(qb_rushing_and_passing_weekly['pfr_player_name']) - set(qb_df['player_display_name']))
# print("Items in list2 but not in list1:", diff2)

# # Checking to see if the list contains only quarterbacks 
# print(qb_rushing_and_passing_weekly['pfr_player_name'].unique())
display(qb_rushing_and_passing_weekly)
print(qb_rushing_and_passing_weekly.columns)
```

```python
qb_snap_counts = nfl.clean_nfl_data(nfl.import_snap_counts(years=[2024]))
qb_snap_counts = qb_snap_counts[qb_snap_counts['position'] == 'QB']
print(qb_snap_counts.columns)

qb_snap_counts = qb_snap_counts.drop(['game_id', 'pfr_game_id', 'game_type','team','opponent','pfr_player_id','season','position'], axis=1)
qb_snap_counts.rename(columns={'player': 'pfr_player_name'}, inplace=True)
display(qb_snap_counts.head())

qb_snap_counts['pfr_player_name'] = qb_snap_counts['pfr_player_name'].replace('Gardner Minshew II', 'Gardner Minshew')

players_to_remove = [
    'Jeff Driskel'
]

# Filter out the specified players
qb_snap_counts = qb_snap_counts[
    ~qb_snap_counts['pfr_player_name'].isin(players_to_remove)
]

# Checking to see if the list contains only quarterbacks 
print(qb_snap_counts['pfr_player_name'].unique()) 



diff2 = list(set(qb_snap_counts['pfr_player_name']) - set(qb_df['player_display_name']))
print("Items in list2 but not in list1:", diff2)

cumalative_df = pd.merge(qb_rushing_and_passing_weekly, qb_snap_counts, on=['pfr_player_name', 'week'], how='outer')
display(cumalative_df)
```

```python
nfl_injuries = nfl.import_injuries(years=[2024])
qb_injuries = nfl_injuries[nfl_injuries['position'] == 'QB']
print(qb_injuries.columns)

qb_injuries = qb_injuries.drop(['season', 'game_type' , 'team', 'gsis_id', 'first_name', 'last_name', 'position'], axis=1)
qb_injuries.rename(columns={'full_name': 'pfr_player_name'}, inplace=True)
display(qb_injuries.head(5))

diff2 = list(set(qb_injuries['pfr_player_name']) - set(qb_df['player_display_name']))
print("Items in list2 but not in list1:", diff2)

players_to_remove = [
   'Jeff Driskel'
]

# Filter out the specified players
qb_injuries = qb_injuries[
    ~qb_injuries['pfr_player_name'].isin(players_to_remove)
]

cumalative_df = pd.merge(cumalative_df, qb_injuries, on=['pfr_player_name', 'week'], how='outer')
display(cumalative_df)
print(cumalative_df["pfr_player_name"].unique())
```

```python
qb_ngs = nfl.clean_nfl_data(nfl.import_ngs_data(stat_type="passing", years=[2024]))
qb_ngs = qb_ngs[qb_ngs['player_position'] == 'QB']
print(qb_ngs.columns)


qb_ngs = qb_ngs.drop(['season', 'season_type', 'team_abbr', 'player_gsis_id', 'player_first_name','player_last_name', 'player_jersey_number', 'player_short_name'], axis=1)
qb_ngs.rename(columns={'player_display_name': 'pfr_player_name'}, inplace=True)

diff2 = list(set(qb_ngs['pfr_player_name']) - set(qb_df['player_display_name']))
print("Items in list2 but not in list1:", diff2)

qb_ngs['pfr_player_name'] = qb_ngs['pfr_player_name'].replace('Joe Milton III', 'Joe Milton')
qb_ngs['pfr_player_name'] = qb_ngs['pfr_player_name'].replace('Michael Penix Jr.', 'Micheal Penix')



display(qb_ngs.head(5))
```
