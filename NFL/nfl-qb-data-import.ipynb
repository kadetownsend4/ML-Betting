```python
import pandas as pd
import nfl_data_py as nfl

# import the weekly data for each player in the 2024 NFL season 
weekly_stats_2024 = nfl.import_weekly_data([2024])

qb_df = weekly_stats_2024[weekly_stats_2024['position'] == 'QB'].copy()

# display the weekly data
display(qb_df.head())
```

```python
# Get weekly passing stats for 2023
weekly_passing = nfl.import_weekly_pfr(s_type="pass", years=[2024])

# Get weekly receiving stats for 2022 and 2023
weekly_receiving = nfl.import_weekly_pfr(s_type="rec", years=[2024])

# Display first few rows
display(weekly_passing.head())
display(weekly_receiving.head())
```

```python
qb_snap_counts = nfl.import_snap_counts(years=[2024]) 
qb_snap_counts = qb_snap_counts[qb_snap_counts['position'] == 'QB']
display(qb_snap_counts.head())
```

```python
nfl_injuries = nfl.import_injuries(years=[2024])
qb_injuries =nfl_injuries[nfl_injuries['position'] == 'QB']
display(qb_injuries.head())
```

```python
qb_ngs = nfl.import_ngs_data(stat_type="passing", years=[2024])
display(qb_ngs.head())
```

```python
nfl.clean_nfl_data(weekly_stats_2024)
nfl.clean_nfl_data(weekly_passing)
nfl.clean_nfl_data(weekly_receiving)
nfl.clean_nfl_data(qb_snap_counts)
nfl.clean_nfl_data(qb_injuries)
nfl.clean_nfl_data(qb_ngs)
```

```python
qb_df.rename(columns={'player_display_name': 'player_merge'}, inplace=True)
print(qb_df.columns)  # Verify the change
weekly_passing.rename(columns={'pfr_player_name': 'player_merge'}, inplace=True)
print(weekly_passing.columns)
merged_df = pd.merge(qb_df, weekly_passing, on=['player_merge', 'week'], how='outer')
display(merged_df)

```
