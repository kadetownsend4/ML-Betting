```python
# Fetch team data
team_rosters = nfl.import_seasonal_rosters([2024])

# I asked ChatGPT for help with filtering the data from the API call
# # https://chatgpt.com/share/67e49261-fcfc-800f-a302-03d2a6125d48 
# Extract the necessary columns
selected_columns = team_rosters[['season', 'team', 'player_id', 'player_name', 'first_name', 'last_name' ,'position', 'jersey_number',
       'status', 'week','game_type']]

# Filter the rows to keep only offensive scoring players
offensive_positions = ['QB', 'TE', 'WR', 'RB', 'K']
filtered_team_rosters = selected_columns[selected_columns['position'].isin(offensive_positions)]

# Mapping of team abbreviations to team IDs
team_id_map = {
    "NYJ": 3430,
    "ARI": 3800,
    "CHI": 810,
    "TEN": 2100,
    "ATL": 200,
    "IND": 2200,
    "BAL": 325,
    "LA": 2510,
    "NYG": 3410,
    "CAR": 750,
    "HOU": 2120,
    "PIT": 3900,
    "GB": 1800,
    "SF": 4500,
    "WAS": 5110,
    "CLE": 1050,
    "KC": 2310,
    "SEA": 4600,
    "MIA": 2700,
    "DAL": 1200,
    "DET": 1540,
    "NO": 3300,
    "TB": 4900,
    "BUF": 610,
    "PHI": 3700,
    "DEN": 1400,
    "LAC": 4400,
    "LV": 2520,
    "NE": 3200,
    "MIN": 3000,
    "JAX": 2250,
    "CIN": 920
}

# Now, create a new column in the filtered data to map team abbreviations to their team IDs
filtered_team_rosters['team_id'] = filtered_team_rosters['team'].map(team_id_map)

# Reorder columns to move 'team_id' after 'team'
column_order = ['season', 'team', 'team_id', 'player_id', 'player_name', 'first_name', 'last_name',
                'position', 'jersey_number', 'status', 'week', 'game_type']

# Reorder the DataFrame
filtered_team_rosters = filtered_team_rosters[column_order]

# Sort the data by player_name and week, so the most recent week comes first
filtered_team_rosters = filtered_team_rosters.sort_values(by=['player_name', 'week'], ascending=[True, False])
# Drop duplicate rows based on player_name, keeping the most recent (first) entry for each player
filtered_team_rosters = filtered_team_rosters.drop_duplicates(subset='player_name', keep='last')

display(filtered_team_rosters['status'].unique())

valid_statuses = ['ACT', 'INA', 'RES']
filtered_team_rosters = filtered_team_rosters[filtered_team_rosters['status'].isin(valid_statuses)]
```

```python
# Drop the 'week' column as it's no longer needed
filtered_team_rosters = filtered_team_rosters.drop(columns=['week'])

# Sort the data by player_name and week, so the most recent week comes first
filtered_team_rosters = filtered_team_rosters.sort_values(by=['team_id', 'player_id'], ascending=[True, False])
# Save to CSV
filtered_team_rosters.to_csv('../Team Data/team_roster_data_filtered.csv', index=False)

# Display the filtered data
display(filtered_team_rosters)
```
