```python
import pandas as pd
print(pd.__version__)
```

```python
import nfl_data_py as nfl

# import the weekly data for each player in the 2024 NFL season 
weekly_stats = nfl.clean_nfl_data(nfl.import_weekly_data([2024]))

# display the weekly data
display(weekly_stats)
```

```python
# Check unique player positions
print(weekly_stats['position'].unique())
```

```python
print(weekly_stats.columns)
```

```python
# Create a filtered copy for WR stats (only receiving-related stats)
rb_df = weekly_stats[weekly_stats['position'].isin(['RB', 'FB'])][['player_id', 'player_name', 'player_display_name', 'position',
       'headshot_url', 'recent_team', 'season', 'week',
       'season_type', 'opponent_team', 'carries', 'rushing_yards',
       'rushing_tds', 'rushing_fumbles', 'rushing_fumbles_lost',
       'rushing_first_downs', 'rushing_epa', 'rushing_2pt_conversions',
       'receptions', 'targets', 'receiving_yards', 'receiving_tds',
       'receiving_fumbles', 'receiving_fumbles_lost', 'receiving_air_yards',
       'receiving_yards_after_catch', 'receiving_first_downs', 'receiving_epa',
       'fantasy_points', 'fantasy_points_ppr']].copy()

# Rename columns for consistency with the RB DataFrame
rb_df.rename(columns={'player_name': 'player_abbr', 'player_display_name': 'player_name'}, inplace=True)

# Display the filtered DataFrame
display(rb_df)

# Fill NaN values with 0
rb_df = rb_df.fillna(0)

# Create a CSV file to store the WR stats
rb_df.to_csv("../weekly_nfl_stats_season/rb_df.csv", index=False)

```

```python
weekly_rushing = nfl.clean_nfl_data(nfl.import_weekly_pfr(s_type="rush", years=[2024]))
weekly_rushing = weekly_rushing.drop(['game_id', 'pfr_game_id', 'game_type','team','opponent','pfr_player_id','season','receiving_broken_tackles', 'carries'], axis=1)
print(weekly_rushing.columns)
weekly_rushing.rename(columns={'pfr_player_name': 'player_name'}, inplace=True)
```

```python
rb_df = pd.merge(rb_df, weekly_rushing,on=['player_name', 'week'], how='left' )
display(rb_df)
```
