```python
import pandas as pd
print(pd.__version__)
```

```python
import nfl_data_py as nfl

# import the weekly data for each player in the 2024 NFL season 
weekly_stats = nfl.clean_nfl_data(nfl.import_weekly_data([2024]))

# display the weekly data
display(weekly_stats)
```

```python
print(weekly_stats.columns)
```

```python
# Create a filtered copy with relevant stats for passing & rushing
qb_df = weekly_stats[weekly_stats['position'] == 'QB'][['player_id', 'player_name', 'player_display_name', 'position',
       'headshot_url', 'recent_team', 'season', 'week',
       'season_type', 'opponent_team', 'completions', 'attempts',
       'passing_yards', 'passing_tds', 'interceptions', 'sacks', 'sack_yards',
       'sack_fumbles', 'sack_fumbles_lost', 'passing_air_yards',
       'passing_yards_after_catch', 'passing_first_downs', 'passing_epa',
       'passing_2pt_conversions', 'pacr', 'dakota', 'carries', 'rushing_yards',
       'rushing_tds', 'rushing_fumbles', 'rushing_fumbles_lost',
       'rushing_first_downs', 'rushing_epa','special_teams_tds', 'fantasy_points', 'fantasy_points_ppr']].copy()

qb_df.rename(columns={'player_name': 'player_abbr', 'player_display_name': 'player_name'}, inplace=True)
display(qb_df)

# create a csv file to store the data
qb_df.to_csv("../weekly_nfl_stats_season/qb_df.csv", index=False)

```

```python
# Get weekly passing stats for 2024
weekly_passing = nfl.clean_nfl_data(nfl.import_weekly_pfr(s_type="pass", years=[2024]))
weekly_passing = weekly_passing.drop(['game_id', 'pfr_game_id', 'game_type','team','opponent','pfr_player_id','season', 'receiving_drop', 'receiving_drop_pct', 'def_times_blitzed', 'def_times_hurried','def_times_hitqb'], axis=1)
# print(weekly_passing.columns)

weekly_rushing = nfl.clean_nfl_data(nfl.import_weekly_pfr(s_type="rush", years=[2024]))
weekly_rushing = weekly_rushing.drop(['game_id', 'pfr_game_id', 'game_type','team','opponent','pfr_player_id','season','receiving_broken_tackles'], axis=1)
# print(weekly_rushing.columns)

qb_rushing_and_passing_weekly = pd.merge(weekly_passing, weekly_rushing, on=['pfr_player_name', 'week'], how='inner')
qb_rushing_and_passing_weekly.rename(columns={'pfr_player_name': 'player_name'}, inplace=True)
display(qb_rushing_and_passing_weekly)
```

```python
qb_snap_counts = nfl.clean_nfl_data(nfl.import_snap_counts(years=[2024]))
qb_snap_counts = qb_snap_counts[qb_snap_counts['position'] == 'QB']
qb_snap_counts = qb_snap_counts.drop(['pfr_game_id', 'game_type','team','opponent','pfr_player_id','season','position', 'defense_snaps', 'defense_pct', 'st_snaps', 'st_pct'], axis=1)
qb_snap_counts.rename(columns={'player': 'player_name'}, inplace=True)
print(qb_snap_counts.columns)
display(qb_snap_counts)
```

```python
qb_ngs = nfl.clean_nfl_data(nfl.import_ngs_data(stat_type="passing", years=[2024]))
qb_ngs = qb_ngs[qb_ngs['player_position'] == 'QB']
qb_ngs.rename(columns={'player_display_name': 'player_name'}, inplace=True)


# Drop Week 0 and Week 1
qb_ngs = qb_ngs[~qb_ngs['week'].isin([0])]

# Drop unneeded columns
qb_ngs = qb_ngs.drop([
    'season', 'season_type', 'team_abbr', 'player_gsis_id',
    'player_first_name', 'player_last_name', 'player_jersey_number',
    'player_short_name', 'attempts',
    'pass_yards', 'pass_touchdowns', 'interceptions',  'completions'
], axis=1) 
```
