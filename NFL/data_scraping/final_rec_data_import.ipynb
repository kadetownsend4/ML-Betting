```python
import pandas as pd
print(pd.__version__)
```

```python
import nfl_data_py as nfl

# import the weekly data for each player in the 2024 NFL season 
weekly_stats = nfl.clean_nfl_data(nfl.import_weekly_data([2024]))

# display the weekly data
display(weekly_stats)
```

```python
print(weekly_stats.columns)
```

```python
# Create a filtered copy for WR stats (only receiving-related stats)
wr_df = weekly_stats[weekly_stats['position'].isin(['WR', 'TE'])][['player_id', 'player_name', 'player_display_name', 'position',
       'headshot_url', 'recent_team', 'season', 'week',
       'season_type', 'opponent_team', 'receptions', 'targets', 'receiving_yards', 'receiving_tds',
       'receiving_fumbles', 'receiving_fumbles_lost', 'receiving_air_yards',
       'receiving_yards_after_catch', 'receiving_first_downs', 'receiving_epa',
       'receiving_2pt_conversions', 'racr', 'target_share', 'air_yards_share',
       'wopr', 'special_teams_tds', 'fantasy_points', 'fantasy_points_ppr']].copy()

# Rename columns for consistency with the WR DataFrame
wr_df.rename(columns={'player_name': 'player_abbr', 'player_display_name': 'player_name'}, inplace=True)

# Display the filtered DataFrame
display(wr_df)

# Create a CSV file to store the WR stats
wr_df.to_csv("../weekly_nfl_stats_season/wr_df.csv", index=False)
```

```python
weekly_recieving = nfl.clean_nfl_data(nfl.import_weekly_pfr(s_type="rec", years=[2024]))
weekly_recieving = weekly_recieving.drop(['pfr_game_id', 'game_type','team','opponent','pfr_player_id','season','rushing_broken_tackles', "passing_drops", "passing_drop_pct"], axis=1)
weekly_recieving.rename(columns={'pfr_player_name': 'player_name'}, inplace=True)
display(weekly_recieving)
```

```python
diff2 = list(set(weekly_recieving['player_name']) - set(wr_df['player_name']))
print("Items in list2 but not in list1:", diff2)
```

```python
wr_ngs = nfl.clean_nfl_data(nfl.import_ngs_data(stat_type="receiving", years=[2024]))
wr_ngs = wr_ngs[wr_ngs['player_position'].isin(['WR', 'TE'])]
wr_ngs.rename(columns={'player_display_name': 'player_name'}, inplace=True)

# Drop Week 0 
wr_ngs = wr_ngs[~wr_ngs['week'].isin([0])]

print(wr_ngs.columns)

# Drop unneeded columns
wr_ngs = wr_ngs.drop([
    'season', 'season_type', 'team_abbr', 'player_gsis_id',
    'player_first_name', 'player_last_name', 'player_jersey_number',
    'player_short_name', 'receptions', 'targets',
    'catch_percentage', 'yards', 'rec_touchdowns', 'player_position'
], axis=1) 

display(wr_ngs)
```

```python
merged_df = pd.merge(wr_df, weekly_recieving, on=['player_name', 'week'], how='left')

merged_df = merged_df.sort_values(by=['player_name', 'week'])

display(merged_df)
```

```python
merged_df = pd.merge(merged_df, wr_ngs, on=['player_name', 'week'], how='left')

merged_df = merged_df.sort_values(by=['player_name', 'week'])

display(merged_df)
```

```python
priority_cols = ['player_id', 'player_name', 'player_abbr', 'game_id', 'week', 'recent_team', 'opponent_team', 'season', 'season_type']
merged_df = merged_df[priority_cols + [col for col in merged_df.columns if col not in priority_cols]]

# Drop rows where game_id is NaN
merged_df = merged_df.dropna(subset=['game_id'])
merged_df = merged_df.dropna()

print(merged_df.columns)
display(merged_df)

# Create a CSV file to store the WR stats
merged_df.to_csv("../weekly_nfl_stats_season/complete_wr_df.csv", index=False)
```
