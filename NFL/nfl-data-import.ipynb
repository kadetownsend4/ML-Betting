# NFL Data Import 
This data is being pulled using [NFL-DATA-PY](https://pypi.org/project/nfl-data-py/) API

```python
import nfl_data_py as nfl

weekly_stats_2024 = nfl.import_weekly_data([2024])

display(weekly_stats_2024)

weekly_stats_2024.to_csv("weekly_stats_2024.csv", index=False)
```
