#### Quarterbacks
Some of the most important factors to consider when analyzing quarter back performance according to [this article](https://mfootballanalytics.com/2020/04/06/which-qb-stats-are-the-most-important/) are:
 - Completion Percentage 
 - Yards per Attempt 
 - Touchdown Percentage 
 - Interception Percentage 
 - Passer Rating 
 - Adjusted Net Yards per Pass Attempt 

```python
# This code block was taken directly from the chat linked above. I decided on the filtered statistics to be used in the dataframe once it gave me 
# guidance on how to filter the dataframe. Then, I asked for help implementing the above statistics into the dataframe. It helped me implement 
# these statistics and suggested additional ones which were the rolling averages and rushing metrics. It provided the code for this as well. 
# Link to chat: https://chatgpt.com/share/67a7d9ef-d054-800f-a70d-36bf42bdf994
# (This is the same link provided above in the markdown comment, but I included it here as well to ensure that the referencing of AI usage is clear)

import numpy as np
import pandas as pd


# Create a filtered copy with relevant stats for passing & rushing
qb_df = weekly_stats_2024[weekly_stats_2024['position'] == 'QB'][['player_id', 'player_name', 'position', 'recent_team', 'week', 'season_type',
                           'opponent_team', 'completions', 'attempts', 'passing_yards',
                           'passing_tds', 'interceptions', 'sacks', 'sack_yards',
                           'carries', 'rushing_yards', 'rushing_tds']].copy()

# Sort by player and week (ensures rolling averages work correctly)
qb_df = qb_df.sort_values(by=['player_id', 'week'])

# Avoid division by zero
qb_df['attempts'] = qb_df['attempts'].replace(0, np.nan)
qb_df['carries'] = qb_df['carries'].replace(0, np.nan)

# Passing Metrics
qb_df['completion_pct'] = (qb_df['completions'] / qb_df['attempts']) * 100
qb_df['yards_per_attempt'] = qb_df['passing_yards'] / qb_df['attempts']
qb_df['td_pct'] = (qb_df['passing_tds'] / qb_df['attempts']) * 100
qb_df['int_pct'] = (qb_df['interceptions'] / qb_df['attempts']) * 100

# Passer Rating Formula
qb_df['a'] = ((qb_df['completions'] / qb_df['attempts']) - 0.3) * 5
qb_df['b'] = ((qb_df['passing_yards'] / qb_df['attempts']) - 3) * 0.25
qb_df['c'] = (qb_df['passing_tds'] / qb_df['attempts']) * 20
qb_df['d'] = 2.375 - ((qb_df['interceptions'] / qb_df['attempts']) * 25)

qb_df[['a', 'b', 'c', 'd']] = qb_df[['a', 'b', 'c', 'd']].clip(0, 2.375)
qb_df['passer_rating'] = ((qb_df['a'] + qb_df['b'] + qb_df['c'] + qb_df['d']) / 6) * 100

# Adjusted Net Yards per Attempt (ANY/A)
qb_df['adjusted_net_yards_per_attempt'] = (
    (qb_df['passing_yards'] - qb_df['sack_yards'] + (20 * qb_df['passing_tds']) - (45 * qb_df['interceptions'])) 
    / (qb_df['attempts'] + qb_df['sacks'])
)

# Drop temp calculation columns
qb_df.drop(columns=['a', 'b', 'c', 'd'], inplace=True)

# Rushing Metrics
qb_df['yards_per_carry'] = qb_df['rushing_yards'] / qb_df['carries']

# Rolling Averages (Last 3 Games, Last 5 Games, Season Avg)
rolling_windows = [3, 5]  # Last 3 games, last 5 games
rolling_cols = ['completion_pct', 'yards_per_attempt', 'td_pct', 'int_pct',
                'passer_rating', 'adjusted_net_yards_per_attempt', 'rushing_yards', 'yards_per_carry']

# Compute rolling averages for each player
for window in rolling_windows:
    for col in rolling_cols:
        qb_df[f'{col}_rolling_{window}'] = qb_df.groupby('player_id')[col].rolling(window, min_periods=1).mean().reset_index(level=0, drop=True)

# Compute season-to-date averages
for col in rolling_cols:
    qb_df[f'{col}_season_avg'] = qb_df.groupby('player_id')[col].expanding().mean().reset_index(level=0, drop=True)

# Save to CSV
qb_df.to_csv("qb_advanced_stats_rolling_2024.csv", index=False)

# # Display relevant columns
# print(qb_df[['player_name', 'week', 'completion_pct', 'yards_per_attempt', 'td_pct', 'passer_rating', 
#              'adjusted_net_yards_per_attempt', 'rushing_yards', 'yards_per_carry', 
#              'completion_pct_rolling_3', 'completion_pct_rolling_5', 'completion_pct_season_avg']].head())
```

#### Recievers
Some of the most important factors to consider when analyzing a recievers performance according to [this article](https://www.sharpfootballanalysis.com/fantasy/wide-receiver-stats-that-matter-fantasy-football-2023/#:~:text=Touchdowns%20are%20less%20important%20for,share%20are%20the%20stickiest%20metrics) are:
 - Receiving yards	
 - Receptions	
 - Targets	
 - Receiving Touchdowns	
 - Snaps

```python
# Filter WRs and select relevant columns
wr_df = weekly_stats_2024[weekly_stats_2024['position'] == 'WR'][[
    'player_id', 'player_name', 'position', 'recent_team', 'week', 'season_type',
    'opponent_team', 'receptions', 'targets', 'receiving_yards', 'receiving_tds',
    'receiving_fumbles', 'receiving_fumbles_lost', 'receiving_air_yards',
    'receiving_yards_after_catch', 'receiving_first_downs', 'receiving_epa',
    'receiving_2pt_conversions', 'racr', 'target_share', 'air_yards_share',
    'wopr', 'fantasy_points', 'fantasy_points_ppr'
]].copy()

# Sort by player and week for rolling averages
wr_df = wr_df.sort_values(by=['player_id', 'week'])

# Avoid division by zero
wr_df.replace(0, np.nan, inplace=True)

### Per-Game Metrics**
wr_df['targets_per_game'] = wr_df['targets']
wr_df['receptions_per_game'] = wr_df['receptions']
wr_df['receiving_yards_per_game'] = wr_df['receiving_yards']
wr_df['receiving_tds_per_game'] = wr_df['receiving_tds']

### Efficiency Metrics**
wr_df['catch_pct'] = (wr_df['receptions'] / wr_df['targets']) * 100  # Catch percentage
wr_df['yards_per_catch'] = wr_df['receiving_yards'] / wr_df['receptions']  # Yards per reception
wr_df['yards_per_target'] = wr_df['receiving_yards'] / wr_df['targets']  # Yards per target
wr_df['air_yards_per_target'] = wr_df['receiving_air_yards'] / wr_df['targets']  # Air yards per target
wr_df['yac_per_reception'] = wr_df['receiving_yards_after_catch'] / wr_df['receptions']  # Yards after catch per reception
wr_df['td_per_target'] = wr_df['receiving_tds'] / wr_df['targets']  # TD conversion rate
wr_df['first_down_per_target'] = wr_df['receiving_first_downs'] / wr_df['targets']  # First downs per target

### Rolling Averages (Last 3, 5 Games & Season-to-Date)
rolling_windows = [3, 5]  
rolling_cols = ['targets_per_game', 'receptions_per_game', 'receiving_yards_per_game', 
                'receiving_tds_per_game', 'catch_pct', 'yards_per_target', 
                'air_yards_per_target', 'yac_per_reception', 'td_per_target', 
                'first_down_per_target']

for window in rolling_windows:
    for col in rolling_cols:
        wr_df[f'{col}_rolling_{window}'] = (
            wr_df.groupby('player_id')[col].rolling(window, min_periods=1).mean().reset_index(level=0, drop=True)
        )

# Season-to-date averages
for col in rolling_cols:
    wr_df[f'{col}_season_avg'] = (
        wr_df.groupby('player_id')[col].expanding().mean().reset_index(level=0, drop=True)
    )

# Save to CSV
wr_df.to_csv("wr_advanced_stats_rolling_2024.csv", index=False)

# # Display relevant columns
# print(wr_df[['player_name', 'week', 'targets_per_game', 'receptions_per_game', 
#              'receiving_yards_per_game', 'receiving_tds_per_game', 'catch_pct', 
#              'yards_per_target', 'air_yards_per_target', 'yac_per_reception',
#              'targets_per_game_rolling_3', 'targets_per_game_rolling_5', 'targets_per_game_season_avg']].head())
```

#### Running Backs 
Some of the most important factors to consider when analyzing quarter back performance according to [this article](https://www.sharpfootballanalysis.com/fantasy/running-back-stats-that-matter-fantasy-football-2023/):
 - Touches	
 - Total Touchdowns	
 - Rushing Attempts	
 - Rushing Yards	
 - Receptions	
 - Rushing TDs	
 - Receiving yards	
 - Snaps	
 - Targets	
 - Receiving TDs


```python
# Filter RBs and select relevant columns
rb_df = weekly_stats_2024[weekly_stats_2024['position'] == 'RB'][[
    'player_id', 'player_name', 'position', 'recent_team', 'week', 'season_type',
    'opponent_team', 'carries', 'rushing_yards', 'rushing_tds',
    'rushing_fumbles', 'rushing_fumbles_lost', 'rushing_first_downs', 'rushing_epa',
    'rushing_2pt_conversions', 'receiving_yards', 'receptions', 'targets',
    'receiving_tds', 'fantasy_points', 'fantasy_points_ppr'
]].copy()

# Sort by player and week for rolling averages
rb_df = rb_df.sort_values(by=['player_id', 'week'])

# Avoid division by zero
rb_df.replace(0, np.nan, inplace=True)

### Per-Game Metrics
rb_df['carries_per_game'] = rb_df['carries']
rb_df['rush_yards_per_game'] = rb_df['rushing_yards']
rb_df['rush_tds_per_game'] = rb_df['rushing_tds']
rb_df['targets_per_game'] = rb_df['targets']
rb_df['receptions_per_game'] = rb_df['receptions']
rb_df['receiving_yards_per_game'] = rb_df['receiving_yards']
rb_df['total_yards_per_game'] = rb_df['rushing_yards'] + rb_df['receiving_yards']  # YFS (Yards from Scrimmage)
rb_df['touches_per_game'] = rb_df['carries'] + rb_df['receptions']  # Total touches

### Efficiency Metrics
rb_df['yards_per_carry'] = rb_df['rushing_yards'] / rb_df['carries']  # Rushing efficiency
rb_df['yards_per_touch'] = (rb_df['rushing_yards'] + rb_df['receiving_yards']) / rb_df['touches_per_game']  # Yards per total touch
rb_df['yards_per_catch'] = rb_df['receiving_yards'] / rb_df['receptions']  # Receiving yards per catch
rb_df['yards_per_target'] = rb_df['receiving_yards'] / rb_df['targets']  # Receiving efficiency per target
rb_df['td_per_touch'] = (rb_df['rushing_tds'] + rb_df['receiving_tds']) / rb_df['touches_per_game']  # TD efficiency

### Rolling Averages (Last 3, 5 Games & Season-to-Date)**
rolling_windows = [3, 5]
rolling_cols = ['carries_per_game', 'rush_yards_per_game', 'rush_tds_per_game',
                'targets_per_game', 'receptions_per_game', 'receiving_yards_per_game',
                'total_yards_per_game', 'touches_per_game', 'yards_per_carry',
                'yards_per_touch', 'yards_per_catch', 'yards_per_target', 'td_per_touch']

for window in rolling_windows:
    for col in rolling_cols:
        rb_df[f'{col}_rolling_{window}'] = (
            rb_df.groupby('player_id')[col].rolling(window, min_periods=1).mean().reset_index(level=0, drop=True)
        )

# Season-to-date averages
for col in rolling_cols:
    rb_df[f'{col}_season_avg'] = (
        rb_df.groupby('player_id')[col].expanding().mean().reset_index(level=0, drop=True)
    )

# Save to CSV
rb_df.to_csv("rb_advanced_stats_rolling_2024.csv", index=False)

# # Display relevant columns
# print(rb_df[['player_name', 'week', 'carries_per_game', 'rush_yards_per_game', 
#              'rush_tds_per_game', 'targets_per_game', 'receptions_per_game', 
#              'receiving_yards_per_game', 'total_yards_per_game', 'touches_per_game', 
#              'yards_per_carry', 'yards_per_touch', 'td_per_touch',
#              'carries_per_game_rolling_3', 'carries_per_game_rolling_5', 
#              'carries_per_game_season_avg']].head())

```
